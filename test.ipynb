```python
import csv

number_exp = "002"

# DATA OF COORD WORD SELECTED
# dictionnaire pour stocker les coordonnées et dimensions des mots sélectionnées
words_coord = {
    "coord_erreurs_1_s" : [],
    "coord_erreurs_2_s" : [],
    "coord_erreurs_3_s" : [],
    "coord_erreurs_4_s" : [],
    "coord_erreurs_1_c" : [],
    "coord_erreurs_2_c" : [],
    "coord_erreurs_3_c" : [],
    "coord_erreurs_4_c" : []
}

coord = []
# Spécifiez le chemin du fichier CSV à lire
chemin_fichier_coord = "donnees_exp_" + number_exp + "\df_coord.csv"

# Ouvrir le fichier en mode lecture
with open(chemin_fichier_coord, 'r') as fichiercoord:
    lecteur_csv = csv.reader(fichiercoord)
    # Parcourir les lignes du fichier CSV
    for indice, line in enumerate(lecteur_csv):
        if indice == 0:  
            continue #en-tête skiped
        
        for idx, _ in enumerate(words_coord):
            if (len(eval(line[idx + 1]))) == 0 :
                words_coord[_].append(eval(line[idx + 1]))
            else :
                for i in range (len(eval(line[idx + 1]))):
                    coord.append([eval(line[idx + 1])[i][0] + eval(line[idx + 1])[i][2]/2, eval(line[idx + 1])[i][0] - eval(line[idx + 1])[i][2]/2, 
                                eval(line[idx + 1])[i][1] + eval(line[idx + 1])[i][3]/2, eval(line[idx + 1])[i][1] - eval(line[idx + 1])[i][3]/2
                                ])
                words_coord[_].append(coord)
            coord = []
        
print('words_coord["coord_erreurs_1_s"] : ', words_coord["coord_erreurs_1_s"])
print('words_coord["coord_erreurs_2_s"] : ', words_coord["coord_erreurs_2_s"]) 
print('words_coord["coord_erreurs_3_s"] : ', words_coord["coord_erreurs_3_s"])
print('words_coord["coord_erreurs_4_s"] : ', words_coord["coord_erreurs_4_s"]) 
print('words_coord["coord_erreurs_1_c"] : ', words_coord["coord_erreurs_1_c"])
print('words_coord["coord_erreurs_2_c"] : ', words_coord["coord_erreurs_2_c"]) 
print('words_coord["coord_erreurs_3_c"] : ', words_coord["coord_erreurs_3_c"])
print('words_coord["coord_erreurs_4_c"] : ', words_coord["coord_erreurs_4_c"]) 
```
